```python
# Извлечение и вывод на экран всех header тегов

from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen("https://en.wikipedia.org/wiki/Main_Page")
soup = BeautifulSoup(html.read(), 'html.parser')
res = [str(tag) for tag in soup.find_all('header')]
print(res)
```

```python
# Скрейпим!

import requests
import random
from bs4 import BeautifulSoup
import string

url="https://www.imdb.com/chart/top"

def getMovies(url):
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/5.140.86.25 '}
  html=requests.get(url,headers=headers)
  soup=BeautifulSoup(html.text,"html.parser")
  movies=soup.find_all("a",class_="ipc-title-link-wrapper")
  random.shuffle(movies)
  return movies

def getSummary(url):
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/5.140.86.25 '}
  movie_sumary=requests.get(url,headers=headers)
  s=BeautifulSoup(movie_sumary.text,"html.parser")
  summary=s.find_all("span",class_="sc-466bb6c-0 kJJttH")
  for x in summary:
        return x.contents[0]

def getMovieInfo(movie):
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/5.140.86.25 '}
  movie_title=movie.contents[0]
  movie_url="https://www.imdb.com"+str(movie.attrs['href'])
  resp=requests.get(movie_url,headers=headers)
  s=BeautifulSoup(resp.text,"html.parser")
  movie_year=s.find_all("a",class_="ipc-link ipc-link--baseAlt ipc-link--inherit-color")
  spisok=[]
  for year in movie_year:
    spisok.append(year.contents[0].text)
  year=spisok[len(spisok)-2]
  return movie_title.text, movie_url ,year

def MoviePicker():
 c=0
 print("--------------------------------------------")
 for movie in getMovies("https://www.imdb.com/chart/top"):
  movie_title, movie_url, movie_year=getMovieInfo(movie)
  print(movie_title, "(",movie_year,")")
  print(getSummary(movie_url))
  print("--------------------------------------------")
  c=c+1
  if (c==10):
          break

if __name__=='__main__':
  MoviePicker()



```
